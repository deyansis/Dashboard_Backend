```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
import pickle
```

```python
data = {
    "comentario": [
        "Buen servicio",
        "Mala atención",
        "Regular",
        "Excelente trabajo",
        "Pésima gestión"
    ],
    "sentimiento": [
        "positivo",
        "negativo",
        "neutral",
        "positivo",
        "negativo"
    ]
}

df = pd.DataFrame(data)
df
```

```python
X = df["comentario"]
y = df["sentimiento"]
```

```python
vectorizer = CountVectorizer()
X_vec = vectorizer.fit_transform(X)
```

```python
modelo = MultinomialNB()
modelo.fit(X_vec, y)
```

```python
texto_prueba = ["La municipalidad trabaja bien"]
texto_vec = vectorizer.transform(texto_prueba)

prediccion = modelo.predict(texto_vec)
print(prediccion)
```

```python
modelo.predict(vectorizer.transform(["Buen servicio"]))
```

```python
import pickle

pickle.dump((modelo, vectorizer), open("modelo.pkl", "wb"))
```

```python
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB

data = {
    "comentario": [
        # positivos
        "Buen servicio",
"Muy buen servicio",
"Excelente servicio",
"Servicio eficiente",
"Servicio de calidad",
"Buena atención",
"Muy buena atención",
"Excelente atención",
        "Al fin se está logrando con la construcción del puente de pisquillo. Siempre habrá críticas porque no a todos les parece bien. Felicitaciones sr alcalde y alos Pobladores de pisquillo por este logro hecho realidad.",
        "Por ahora magnifico",
        "ME PARECE QUE LOS HECHOS DARAN CREDIBILIDAD A LA GESTIÓN, POR AHORA MAGNÍFICO, SALUDOS VICTOR UN GRAN ABRAZOTE... SALUDOS FRATERNOS A TODA MI LINDA GENTE DE PISQUILLO DIOS PADRE TODOPODEROSO ETERNO Y VERDADERO OS BENDIGA AMÉN.UN GRAN ABRAZOTE A CADA UNO HERMANOS",

        # negativos
        "Mala gestión municipal",
        "No hacen nada por el pueblo",
        "Pésima atención",
        "Todo es mentira",
        "No cumplen lo que prometen",
        "La municipalidad no responde",
        "Hay mucha demora",
        "Mal trabajo del alcalde",
        "No solucionan los problemas",
        "Solo quieren llevarse el crédito",

        # neutrales
        "Servicio regular",
        "Atención normal",
        "Todo sigue igual",
        "Sin novedades",
        "Información publicada",
        "Se realizó una actividad",
        "La obra continúa en proceso",
        "Comentario informativo",
        "El comunicado fue publicado",
        "No esta ni bien ni mal"
    ],
    "sentimiento": [
        # positivos
        "positivo", "positivo", "positivo", "positivo", "positivo",
        "positivo", "positivo", "positivo", "positivo", "positivo",
        "positivo",
        

        # negativos
        "negativo", "negativo", "negativo", "negativo", "negativo",
        "negativo", "negativo", "negativo", "negativo", "negativo",

        # neutrales
        "neutral", "neutral", "neutral", "neutral", "neutral",
        "neutral", "neutral", "neutral", "neutral", "neutral"
    ]
}

df = pd.DataFrame(data)

X = df["comentario"]
y = df["sentimiento"]

vectorizer = TfidfVectorizer(ngram_range=(1, 2), lowercase=True)
X_vec = vectorizer.fit_transform(X)

modelo = MultinomialNB()
modelo.fit(X_vec, y)

pruebas = [
    "No está mal ni bien",
    "El servicio fue regular",
    "La atención fue normal",
    "Buen servicio",
    "Pésima gestión",
    "ME PARECE QUE LOS HECHOS DARAN CREDIBILIDAD A LA GESTIÓN, POR AHORA MAGNÍFICO, SALUDOS VICTOR UN GRAN ABRAZOTE... SALUDOS FRATERNOS A TODA MI LINDA GENTE DE PISQUILLO DIOS PADRE TODOPODEROSO ETERNO Y VERDADERO OS BENDIGA AMÉN.UN GRAN ABRAZOTE A CADA UNO HERMANOS"
    "Al fin se está logrando con la construcción del puente de pisquillo. Siempre habrá críticas porque no a todos les parece bien. Felicitaciones sr alcalde y alos Pobladores de pisquillo por este logro hecho realidad.",
        "Por ahora magnifico",
]

for texto in pruebas:
    pred = modelo.predict(vectorizer.transform([texto]))[0]
    print(texto, "=>", pred)

pickle.dump((modelo, vectorizer), open("modelo.pkl", "wb"))
```

```python
import pandas as pd

from sklearn.feature_extraction.text import TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB

import pickle
```

```python
data_filtro = {

    "texto": [

        "Pésima gestión",
        "Excelente iniciativa",
        "Muy buena atención",
        "La municipalidad debería mejorar",

        "Número de notificaciones no leídas",
        "Crear cuenta",
        "Messenger",
        "Compartir",
        "Meta",
    ],

    "tipo": [

        "valido",
        "valido",
        "valido",
        "valido",

        "ruido",
        "ruido",
        "ruido",
        "ruido",
        "ruido",
    ]
}
```

```python
df_filtro = pd.DataFrame(data_filtro)

df_filtro.head()
```

```python
X_filtro = df_filtro["texto"]

y_filtro = df_filtro["tipo"]
```

```python
vectorizer_filtro = TfidfVectorizer()

X_filtro_vec = vectorizer_filtro.fit_transform(
    X_filtro
)
```

```python
modelo_filtro = MultinomialNB()

modelo_filtro.fit(
    X_filtro_vec,
    y_filtro
)
```

```python
pruebas_filtro = [

    "Número de notificaciones no leídas",

    "Excelente trabajo municipal",

    "Messenger",

    "Pésima gestión"
]

for texto in pruebas_filtro:

    pred = modelo_filtro.predict(

        vectorizer_filtro.transform(
            [texto]
        )

    )[0]

    print(texto, "=>", pred)
```

```python
pickle.dump(

    (modelo_filtro, vectorizer_filtro),

    open(
        "modelo_filtro.pkl",
        "wb"
    )
)
```
